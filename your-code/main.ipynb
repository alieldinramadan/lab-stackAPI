# StackAPI

#### Import the necessary libraries here:

```python
import pandas as pd
import requests
import time
import datetime
from stackapi import StackAPI
```

```python
SITE = StackAPI('stackoverflow')
```

```python
#To get the time stamos for the first and last day in October, se I can pass it in the fetch function (used in question 1)
print(time.mktime(time.strptime('01/10/2019', "%d/%m/%Y")))
print(time.mktime(time.strptime('31/10/2019', "%d/%m/%Y")))
print(time.mktime(time.strptime('07/11/2019', "%d/%m/%Y"))) #To get timestamp for today (used in question 2)
```

#### Question 1: Find the questions and answers of last month.

```python
questions = SITE.fetch('questions', fromdate=1569880800, todate=1572476400) #to get the questions of last month
answers = SITE.fetch('answers', fromdate=1569880800, todate=1572476400) ##to get the answers of last month
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
most_voted_questions = SITE.fetch('questions', fromdate=1573081200, min=5, tagged='python', sort='votes')
most_voted_questions
```
